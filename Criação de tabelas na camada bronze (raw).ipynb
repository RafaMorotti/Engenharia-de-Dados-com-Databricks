```python
# inicio projeto - Git Databricks
# caminho para o arquivo brands
brands_path = ("/Volumes/pipeline_bikes/zone_bronze/volume_bronze/brands.csv")
brands_table_name = ("pipeline_bikes.zone_bronze.brands_bronze")

# carregando o arquivo brands em df_brands
df_brands = (
    spark.read
    .option("header", True)
    .option("inferSchema", True)
    .option("sep", ",")
    .csv(brands_path)
)

# carregando o DF_Brands para tabela, sem alterações camada Bronze (raw)
df_brands.write \
    .format("delta") \
    .mode("overwrite") \
    .saveAsTable(brands_table_name)
```
